```python
weather_df = pd.read_csv("https://djl-lms-assets.s3.eu-central-1.amazonaws.com/datasets/weatherHistory.csv")
weather_df.head()
```

Hedef değişkeninizin görünür sıcaklık ve sıcaklık arasındaki fark olduğu doğrusal bir regresyon modeli oluşturun. Açıklayıcı değişkenler olarak nem ve rüzgar hızı kullanın. Şimdi, modelinizi OLS kullanarak tahmin edin. Tahmin edilen katsayılar istatistiksel olarak anlamlı mıdır? Tahmini katsayılar önceki beklentileriniz doğrultusunda mı? Tahmin edilen katsayıları yorumlayınız. Hedef ve açıklayıcı değişkenler arasındaki ilişkiler nelerdir?

```python
y=weather_df["Apparent Temperature (C)"]-weather_df["Temperature (C)"]
```

```python
x=weather_df[["Humidity","Wind Speed (km/h)"]]
```

```python
import statsmodels.api as sm
x = sm.add_constant(x)
results = sm.OLS(y, x).fit()
results.summary()
```

Tahmin edilen katsayılar istatiksel açıdan p değerlerine baktığımızda anlamlıdır. Rüzgar hızı ve nem ile hava sıcaklığı negatif ilişkili gözükmektedir. bir birim rüzgar ile 0.1193 katı hava düşmektedir. bir birim nem ile 3.0292 katı kadar hava sıcaklığı düşmektedir Sabit ile beraber .

Ardından, yukarıdaki modele nem ve rüzgar hızı etkileşimini dahil edin ve OLS'yi kullanarak modeli tahmin edin. Katsayılar istatistiksel olarak anlamlı mıdır? Nem ve rüzgar hızı için tahmini katsayıların işaretleri değişti mi? Tahmin edilen katsayıları yorumlayınız.

```python
weather_df["nem*ruz"]=weather_df["Humidity"]*weather_df["Wind Speed (km/h)"]
```

```python
x=weather_df[["Humidity","Wind Speed (km/h)","nem*ruz"]]
```

```python
import statsmodels.api as sm
x = sm.add_constant(x)
results = sm.OLS(y, x).fit()
results.summary()
```

p değerlerine bakarak değişkenlerin istatiksel olarak anlamlı olduğunu söyleriz. Bir önceki modele göre işaretleri değişti.
nem ve rüzgar hızı ile pozitif artmakta, ruz hızı*nem etkileşimli değişkenim ile negatif olarak etkilemektedir sıcaklık farkımı.
